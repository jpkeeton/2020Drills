Modes for file open

* r
    * open for reading
    * default
* r+
    * open for reading and writing
    * file pointer is at the beginning of the file
* w
    * open for writing
    * truncate the file if it exists
* w+
    * open for both reading and writing 
    * truncate the file if it exists
* a
    * open for writing
    * append to the end of the file if exists & file pointer is at the end of the file
