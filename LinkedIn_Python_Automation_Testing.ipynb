**XPath**

* Xpath stands for XML Path Language
* Uses path expressions to identify and navigate nodes in an XML document
* HTML is an implementation of XML
* can locate in absolute or relative paths
 * recommended to NOT use absolute path, could change
 * relative is better
 * It's suggested to only use Xpath when there is no id, class or name attribute is not available
 * As Xpath follows a tree or a folder structure, the path expressions resemble those used with traditional computer file systems.
 * As the absolute xpath is specified right from the root node to the element, any changes in the application structure could render it invalid. Hence, it is advised to use relative xpath.

**Drag and Drop**
* https://www.linkedin.com/learning/python-automation-and-testing/drag-and-drop-elements
* As part of a UI interaction
* Source element to target elements
* Source element to specific x/y
* We'll use ActionChains
    ** see https://selenium-python.readthedocs.io/api.html

```python
from selenium import webdriver
import time
# import ActionChains also
from selenium.webdriver import ActionChains
from selenium.webdriver.chrome.options import Options

driver = webdriver.Firefox()
# maximize the window
driver.maximize_window()
# use this site to manipulate stuff
print('Going to jquery site')
driver.get('https://jqueryui.com/droppable/')
# Switch the frame
driver.switch_to.frame(0)

action_chains=ActionChains(driver)

source = driver.find_element_by_id('draggable')
target = driver.find_element_by_id('droppable')


action_chains.drag_and_drop_by_offset(source, 100, 100).perform()
time.sleep(2)
action_chains.drag_and_drop(source, target).perform()

time.sleep(2)




print('Closing browser')
driver.close()
```

**Waits**

* https://www.linkedin.com/learning/python-automation-and-testing/what-are-waits-and-why-do-we-need-them
* Ajax etc can cause elements to load more slowly
* Explicit and Implicit Waits
    * Explicit
        * Stops execution until a certain condition is met
        * Selenium Webdriver uses a combo of the WebDriverWait and ExpectedConditions classes to add explicit waits
        * waits for a certain condition only
    * Implicit
        * Polls the DOM for a given amount of time while trying to locate an element
        * waits for every call

```python
# Explicit Wait Example

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


driver =webdriver.Firefox()
driver.get('http://python.org')
driver.maximize_window()
# add a 10sec wait
# if the element is loaded w/in 10 sec we'll proceed

try:
    element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, 'start-shell'))
    )
finally:
    driver.quit()

 
```

```python
# implicit waits are good for slow connections
from selenium import webdriver
driver =webdriver.Firefox()

driver.implicitly_wait(10)
driver.get('http://python.org')

myDynamicElemeent = driver.find_element_by_id('start-shell')



print('Closing browser')
driver.close()

```
