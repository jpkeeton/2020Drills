**Using 'iloc' to subset**

* integer location, but it's analogous to 'loc'
* can also select rows, columns or both
* syntax = df.iloc[rows, columns]

* 3 Selection types:
    * single integer
    * list of integers
    * slice of integers




```python
# create some list for rows and columns
ilocRows = [2, 4]
ilocColumns = [0, -1]
```

```python
sampleData.iloc[ilocRows, ilocColumns]
```

```python
sampleData
```

```python
# Let's say we want the first and last columns
cols2 = [0, -1]
# then use slice notation (the '':3' bit) and a list (the 'cols2' bit) to get the first 3 rows and the cols2 above
sampleData.iloc[:3, cols2]
```

```python
# Select some rows and ALL the columns, just use slice notation to get the 
# and the iloc bit is Exclusive, will not give you back your last number, and the column option is of course not needed
sampleData.iloc[1:3, :]
```

```python
# Now select some rows and a single column
sampleData.iloc[[2, 3, 4], 2]
# this will return a series
```

```python
# but what if you want some rows and a single column to be returned as a DataFrame?
# just use a single item list
sampleData.iloc[[2, 3, 5],[3]]
```

```python
# Select a single rows as series
sampleData.iloc[3, [3, 2, 1]]
```

```python
# then select one row as a dataframe
sampleData.iloc[[3], [3, 2, 1, -1]]
```

```python
# or one row and all the columns
sampleData.iloc[[3], :]
```
