**Boolean Indexing on multiple conditions**

```python
# create complex filters with the AND &, OR |, NOT ~ operators
# best practice to assign each filter as it's own variable
# use the 'isin' method to test for multiple equalities in the same column

bikes.head(3)
```

```python
# so let's get all rides where it was cloudy and duration was greater than 1000 seconds
weatherFilt = bikes['events'] == 'cloudy'
```

```python
durationFilt = bikes['tripduration'] < 100
```

```python
bikes[durationFilt & weatherFilt].head()
```

```python
bigFilt = durationFilt & weatherFilt
```

```python
len(bigFilt)
```

```python
# you can also filter on multiple conditions in one line, just use parens
newFilt = (bikes['tripduration'] > 1000) & (bikes['events'] =='cloudy')
```

```python
bikes[newFilt].head(2)
```

OR operator | 

```python
# OR operator
newFiltOR = (bikes['gender'] == 'female') | (bikes['tripduration'] < 100)
```

```python
bikes[newFiltOR].head(111)
```

NOT operator ~, aka inverion operator

```python
notFilt1 = bikes['tripduration']> 1000
# Put the tilde here!
# but redundant here, just to see it working
bikes[~notFilt1]
```

ISIN - test multiple conditions

```python
specialEvents = ['rain', 'snow', 'sleet']
filt = bikes['events'].isin(specialEvents)
```

```python
bikes[filt]
```

**Boolean Selection on a Series**
* Using 'between' as opposed to an 'and'
* Using 'isna' method for rows with missing values
* maybe not used as much
* usually you'll do loc or iloc when selecting on a series, but Boolean selection is ok   


```python
# Select the temp column
temp =  bikes['temperature']
temp.head()
```

```python
# Select temps > 90 from a series

filt = temp > 90
filt
```

```python
# Remember you can set your index to whatever, like 'index_col = 'starttime', so make things more useful when needed
```

**Between Method**

```python
# temp.head()
filt = temp.between (50, 60)
filt
```

```python
#Simultaneous Boolean selection for row and column labels with loc
filt = temp.between(50, 60)
cols = ['temperature', 'events']
bikes.loc[filt, cols]
```

```python
# Column to Column Comparisons
bikes['dpcapacity_start'] > bikes['dpcapacity_end']
```

```python
# Finding missing values w isna
bikes['dpcapacity_start'].isna()
```

```python
# then again as a filter
filterIsna = bikes['dpcapacity_start'].isna()
bikes[filterIsna]
```

```python
# # Summary
    # ok to use brackets with a Series
    # use Loc for simultaneous boolean selection and column selection
    # between is a shortcut for And
```

**Query Method**

```python
bikes.query('tripduration > 1000').head()
# simpler than the below...

# bikes[bikes['tripduration'] < 1000]

# doesn't work on columns with spaces!!
```

```python
# and with Booleans! here you can use the word
# bikes.query('tripduration > 1000 and temperature >90')

# another example with the Not
bikes.query('not(tripduration > 1000 and temperature >90)')
```

```python
# @ symbol used w query
td = 1000
bikes.query('tripduration > @td')
```
